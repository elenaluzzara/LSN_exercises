```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def f(x, y):
    return np.sin(x**2 + y**2)
```

## Exercise 11.3

The goal of this last exercise is to extend the model of neural networks to fit a simple trigonometric 2D function such as $f(x,y) = \sin(x^2+y^2)$ in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

```python
np.random.seed(0)
n_train = 10000
n_valid = 1000
sigma = 0.2 # noise standard deviation, for the moment it is absent
low=-3/2
high=3/2

x_train = np.zeros((n_train, 2)) #inpunt (x,y)
x_valid = np.zeros((n_valid, 2)) #inpunt (x,y)

#y_train = np.zeros((n_train, 1)) #inpunt (x,y)
#y_valid = np.zeros((n_valid, 1)) #inpunt (x,y)

x_train[:,0] = np.random.uniform(low, high, n_train) #troppo pochi per una buona predizione, ha imparato poco
x_train[:,1] = np.random.uniform(low, high, n_train)

x_valid[:,0] = np.random.uniform(low, high, n_valid)
x_valid[:,1] = np.random.uniform(low, high, n_valid)

y_train = np.random.normal(np.sin(x_train[:,0]**2 + x_train[:,1]**2), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(np.sin(x_valid[:,0]**2 + x_valid[:,1]**2), sigma)

y_target=f(x_valid[:,0],x_valid[:,1]) 

x1 = np.copy(x_valid[:,0])
x2 = np.copy(x_valid[:,1])
x1.sort()
x2.sort()
x1, x2 = np.meshgrid(x1,x2)
y = f(x1, x2)
```

```python
fig = plt.figure(figsize=(10, 8))
ax = Axes3D(fig)
#ax.contour3D(x1, x2, y_target, 50, cmap='binary')
#ax.scatter3D(x_valid[:,0], x_valid[:,1], y_valid, marker='o', color='red')
ax.scatter(x_valid[:,0], x_valid[:,1], y_valid, c=y_valid, cmap='winter', label='Test')
ax.plot_surface(x1, x2 , y, cmap='GnBu', linewidth=0, antialiased=True, alpha=0.3)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

I trained a model with the same choice of parameters of the previous exercise:
- optimizer: *rmsprop*
- number of layers: $HL=5$
- number of neurons for layer: $N=90$ 
- activation:*relu*
- loss function: *msle*

The result is a good prediction of the model. 

```python
model = tf.keras.Sequential()
model.add(Dense(units=90, input_shape=(2,)))

for i in range(5):
    model.add(Dense(units=90, activation = 'relu'))

model.add(Dense(units=1))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='rmsprop', loss='msle', metrics=['mse']) 
```

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train,
          batch_size=50, epochs=60,
          shuffle=True, # a good idea is to shuffle input before at each epoch cos che non usi sempre gli stessi dati di batch
          validation_data=(x_valid, y_valid))
```

```python
# print performance
score = model.evaluate(x_valid, y_valid, batch_size=50, verbose=1)
print("**Performance**")
print('Test loss:', score[0])

x_predicted = np.zeros((300, 2)) #inpunt (x,y)
x_predicted[:,0] = np.random.uniform(-3/2, 3/2, 300)
x_predicted[:,1] = np.random.uniform(-3/2, 3/2, 300)

y_predicted = model.predict(x_predicted)

fig = plt.figure(figsize=(10,6))
ax = Axes3D(fig)
ax.scatter(x_predicted[:,0], x_predicted[:,1], y_predicted,  c='navy', cmap='winter', label='Test', s=50)
ax.plot_surface(x1, x2 , y, cmap='GnBu', linewidth=0, antialiased=True, alpha=0.3)

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(50, 35)
plt.show()
```
