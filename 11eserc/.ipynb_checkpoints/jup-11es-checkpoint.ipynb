# Exercise 11.1

The task of this assignment is to perform machine learning regression on noisy data with a Neural Network (NN).

This first exercise uses NN to predict through a neural network regression the parameters of the linear function $f(x) = 2x +1$.

```python
# target parameters of f(x) = m*x + b
m = 2 # slope
b = 1 # intersect
```

```python
import numpy as np

# generate training inputs
np.random.seed(0)
n = 100
x_train = np.random.uniform(-1, 1, 10*n) #10troppo pochi per una buona predizione, ha imparato poco
x_valid = np.random.uniform(-1, 1, 100)
x_valid.sort()
y_target = m * x_valid + b # ideal (target) linear function

sigma = 0.5 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(m * x_train + b, sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(m * x_valid + b, sigma)
```

```python
# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(1, input_shape=(1,)))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mse', metrics=['mse'])
```

```python
model.summary()
```

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
n_epochs=80
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=n_epochs,
          shuffle=True, # a good idea is to shuffle input before at each epoch così che non usi sempre gli stessi dati di batch
          validation_data=(x_valid, y_valid))
```

```python
# return weights
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
model_epochs = np.column_stack((history.history['loss'],history.history['val_loss']))
np.savetxt(f"10000model_sigma{sigma}_{n_epochs}epochs.out",model_epochs)
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

How does the linear regression depend on the number of epochs, $N_{\mathrm{epochs}}$, the number of data points $N_{\mathrm{train}}$ and on the noise $\sigma$?

Mainting fixed the other parameters, if sigma is different from 0, changing the number of epochs can improve the loss function and the test accurancy up to a maximum. Over that maximum the loss function doesn't decrease anymore and reaches a stationary value. The higher is the noise $\sigma$ the quicker this value is reached. The number of data points $N_{\mathrm{train}}$ improves the learning of the machine if it increased. 

```python
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

xc,yc = np.loadtxt("10000model_sigma0.3_30epochs.out", usecols =(0,1), unpack = True)
xa,ya = np.loadtxt("10000model_sigma0.5_80epochs.out", usecols =(0,1), unpack = True)
x1 = np.linspace(0,30,30)
x2 = np.linspace(0,80,80)
ax.plot(x1,xc,label = "0.3")
ax.plot(x2,xa, label = "0.5")
ax.set_xlim(0,10)
ax.plot(x1,yc)
ax.plot(x2,ya)
ax.legend(loc=(0.7,0.7))
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

```python
print("Input dataset shape: ", x_valid.shape)
# this model maps a 0 dim (non è un array) problem into an 1 dims (è un array con un elemento)
y_predicted = model.predict(x_valid, batch_size=128)
print("Predicted results shape: ", y_predicted.shape)
print(y_valid[4])
print(y_predicted[4])
```

### Esercizi 2

Try to extend the model to obtain a reasonable fit of the following polynomial of order 3:

$$
f(x)=4-3x-2x^2+3x^3
$$
for $x \in [-1,1]$.

```python
import numpy as np

a = 4 
b = -3 
c = -2
d = 3
# generate training inputs
np.random.seed(0)
x_train = np.random.uniform(-1, 1, 10000) #troppo pochi per una buona predizione, ha imparato poco
x_valid = np.random.uniform(-1, 1, 100)
x_valid.sort()
y_target = a + b * x_valid + c * (x_valid**2) + d * (x_valid**3) # ideal (target) linear function

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(a + b * x_train + c * (x_train**2) + d * (x_train**3), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(a + b * x_valid + c * (x_valid**2) + d * (x_valid**3), sigma)
```

```python
# plot training and target dataset
import matplotlib.pyplot as plt
plt.plot(x_valid, y_target)
plt.scatter(x_valid, y_valid, color='r')
plt.grid(True); plt.show()
```

```python
# compose the NN model
import tensorflow as tf
from tensorflow import keras

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras import backend as K
from tensorflow.keras.utils import get_custom_objects

model = tf.keras.Sequential()
model.add(Dense(units=400, input_shape=(1,)))
#model.add(Activation('relu'))
model.add(Dense(units=45, activation = 'relu'))
#model.add(Activation('relu'))
model.add(Dense(units=1,activation = 'relu'))

# compile the model choosing optimizer, loss and metrics objects
model.compile(optimizer='sgd', loss='mae', metrics=['mse']) 
#mae sembra essere la migliore con Dtrain = 10000 e sigma = 0 e activation = relu e Nepoch = 50
```

```python
model.summary()
```

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train, 
          batch_size=32, epochs=60,
          shuffle=True, # a good idea is to shuffle input before at each epoch così che non usi sempre gli stessi dati di batch
          validation_data=(x_valid, y_valid))
```

```python
model.get_weights()
```

```python
# evaluate model
score = model.evaluate(x_valid, y_valid, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# evaluate model with the exact curve
score = model.evaluate(x_valid, y_target, batch_size=32, verbose=1)

# print performance
print()
print('Test loss:', score[0])
print('Test accuracy:', score[1])

# look into training history
# Plot training & validation loss values
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='best')
plt.show()
```

```python
print("Input dataset shape: ", x_valid.shape)
# this model maps a 0 dim (non è un array) problem into an 1 dims (è un array con un elemento)
y_predicted = model.predict(x_valid, batch_size=128)
print("Predicted results shape: ", y_predicted.shape)
print(y_valid[4])
print(y_predicted[4])
```

```python
x_predicted = np.random.uniform(-1, 1, 100)
y_predicted = model.predict(x_predicted)
plt.scatter(x_predicted, y_predicted,color='r')
plt.plot(x_valid, y_target)
plt.grid(True)
plt.show()
```

## Esercizi 11.3

$f(x,y) = \sin(x^2+y^2)$ in the range $x \in [-3/2,3/2]$ and $y \in [-3/2,3/2]$.

```python
x_train = np.zeros((10, 2)) 
x_train[:,1] = np.random.uniform(-3/2, 3/2, 10)
print(x_train[:,1])
```

```python
import numpy as np

# generate training inputs
np.random.seed(0)
n_train = 10000
n_valid = 100
x_train = np.zeros((n_train, 2)) #inpunt (x,y)
x_valid = np.zeros((n_valid, 2)) #inpunt (x,y)

y_train = np.zeros((n_train, 1)) #inpunt (x,y)
y_valid = np.zeros((n_valid, 1)) #inpunt (x,y)

x_train[:,0] = np.random.uniform(-3/2, 3/2, n_train) #troppo pochi per una buona predizione, ha imparato poco
x_train[:,1] = np.random.uniform(-3/2, 3/2, n_train)

x_valid[:,0] = np.random.uniform(-3/2, 3/2, n_valid)
x_valid[:,1] = np.random.uniform(-3/2, 3/2, n_valid)

#x_valid.sort()
def f(x, y):
    return np.sin(x ** 2 + y ** 2)

x1, x2 = np.meshgrid(x_valid[:,0], x_valid[:,1])
y_target = f(x1, x2)

sigma = 0.0 # noise standard deviation, for the moment it is absent
y_train = np.random.normal(np.sin(x_train[:,0]**2 + x_train[:,1]**2), sigma) # actual measures from which we want to guess regression parameters
y_valid = np.random.normal(np.sin(x_valid[:,0]**2 + x_valid[:,1]**2), sigma)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x1, x2, y_target, 50, cmap='binary')
ax.scatter3D(x_valid[:,0], x_valid[:,1], y_valid, marker='o', color='red')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(60, 35)
plt.show()

```

```python
# fit the model using training dataset
# over 10 epochs of 32 batch size each
# report training progress against validation data
history = model.fit(x=x_train, y=y_train,
          batch_size=32, epochs=60,
          shuffle=True, # a good idea is to shuffle input before at each epoch così che non usi sempre gli stessi dati di batch
          validation_data=(x_valid, y_valid))
```
