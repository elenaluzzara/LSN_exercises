```python
import matplotlib.pyplot as plt
import numpy as np
fig, ax = plt.subplots(1, 1, figsize=(10, 6))

Nepochs_1 = 30
Nepochs_2 = 30
sigma_1 = 0.3
sigma_2 = 0.3
Ntrain_1 = 500
Ntrain_2 = 1000
xtrain_1,ytest_1 = np.loadtxt(f"{Ntrain_1}_model_sigma{sigma_1}_{Nepochs_1}epochs.out", usecols =(0,1), unpack = True)
xtrain_2,ytest_2 = np.loadtxt(f"{Ntrain_2}_model_sigma{sigma_2}_{Nepochs_2}epochs.out", usecols =(0,1), unpack = True)
x1 = np.linspace(0,Nepochs_1,Nepochs_1)
x2 = np.linspace(0,Nepochs_2, Nepochs_2)
ax.plot(x1,xtrain_1,label = f"Train - sigma {sigma_1} - Ntrain {Ntrain_1} - Nepochs {Nepochs_1}", color = 'red')
ax.plot(x2,xtrain_2, label = f"Train - sigma {sigma_2} - Ntrain {Ntrain_2} - Nepochs {Nepochs_2}", color = 'blue')
ax.set_xlim(0,30)
ax.plot(x1,ytest_1, label = f"Test - sigma {sigma_1} - Ntrain {Ntrain_1} - Nepochs {Nepochs_1}", color = 'red', alpha = 0.5)
ax.plot(x2,ytest_2, label = f"Test - sigma {sigma_2} - Ntrain {Ntrain_2} - Nepochs {Nepochs_2}", color = 'blue', alpha = 0.5)
ax.legend(loc=(0.3,0.7))
```
