```python
energy = np.loadtxt("instant_epot.out", usecols=(0), unpack='true')
pression = np.loadtxt("instant_pres.out", usecols=(0), unpack='true')

prodotto = 0
somma = 0
somma2 = 0
somma_parz = 0
somma_parz2 = 0

def funz_autocorr(valore, passo, dime):
    global prodotto 
    global somma 
    global somma2 
    global somma_parz
    global somma_parz2
    for i in range(dime - passo):
        prodotto += valore[i]*valore[i+passo]
        somma_parz += valore[i]
        somma_parz2 += valore[i+passo]
    for i in range(dime):
        somma += valore[i]
        somma2 += pow(valore[i],2)
    prodotto /=(dime - passo)
    somma_parz /=(dime - passo)
    somma_parz2 /=(dime - passo)
    somma2 /=(dime)
    devstd = (somma2-pow(somma/(dime),2))
    return (prodotto - somma_parz*somma_parz2)/devstd
    
energy_autocorr = np.zeros(100)
pression_autocorr = np.zeros(100)
    
for t in range(100):
    energy_autocorr[t] = funz_autocorr(energy,t, 500000)
    pression_autocorr[t] = funz_autocorr(pression,t,500000)
```

```python
x = np.linspace(0,99, num=100)
plt.plot(x,energy_autocorr)
```

```python
def error(AV,AV2,n):
    if n==0: 
        return 0
    else:
        return math.sqrt((AV2[n] - AV[n]**2)/n) 

def block_average(M,L,valore):
    N=int(M/L)            
    x = np.arange(N)      
    ave = np.zeros(N)
    av2 = np.zeros(N)
    sum_prog = np.zeros(N)
    su2_prog = np.zeros(N)
    err_prog = np.zeros(N)

    for i in range(N): 
        sum = 0
        for j in range(L): 
            k = j+i*L 
            sum += valore[k]
        ave[i] = sum/L      
        av2[i] = (ave[i])**2 

    for i in range(N): 
        for j in range(i+1): #ciclo interno che va a sommare gli esperimenti fino a quel momento
            sum_prog[i] += ave[j] # SUM_{j=0,i} r_j (accumolo valor medio)
            su2_prog[i] += av2[j] # SUM_{j=0,i} (r_j)^2
        sum_prog[i]/=(i+1) # Cumulative average
        su2_prog[i]/=(i+1) # Cumulative square average
        err_prog[i] = error(sum_prog,su2_prog,i) # Statistical uncertainty
    x*=L 
    
    fig, ax = plt.subplots(1, 1, figsize=(10, 6))
    ax.errorbar(x,sum_prog,yerr=err_prog)
    plt.xlabel('nstep')
    plt.ylabel('valore')
    plt.grid(True)
    plt.show()
```
